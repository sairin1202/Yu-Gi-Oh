```python
import requests
import urllib.request
from bs4 import BeautifulSoup
import os
import time
import re
import cv2
```

```python
count=0

folder_path = './raw/'
if os.path.exists(folder_path) == False:  # 判断文件夹是否已经存在
    os.makedirs(folder_path)  # 创建文件夹
```

```python
for i in range(769):
    print("index",i)
    url  = "http://www.cardbox.sc/cardlist_ygo?exs=&exp=&rrt=&kind=&trb=&attr=&llv=&hlv=&lo=&ho=&ld=&hd=&lmt=&word=&offset="+str(i)
    response = requests.get(url) 
    #print(response.content.decode('utf-8'))
    items = re.findall('http://www.cardbox.sc/img/card/ygo/.+\.jpg|/img/card/noimage.jpg|★\d+',response.content.decode('utf-8'),re.I)
    items.append("end")
    print(items)
    extracted_items=[items[i] for i in range(len(items)-1) if (items[i][0]=='★' or items[i+1][0]=='★') and ("noimage" not in items[i] and ("noimage" not in items[i-1] or items[i][0]!='★'))]
    print(extracted_items)

    for index,item in enumerate(extracted_items):
        if index%2==0:
            html = requests.get(item)  
            img_name = folder_path + str(count) +'.jpg'
            with open(img_name, 'wb') as file:  
                file.write(html.content)
                file.flush()
            file.close()
            print('%d images have been downlaoded' %(count+1))
            time.sleep(0.2) 
            #img = cv2.imread("./dataset/"+str(count)+".jpg")
            #img = cv2.resize(img,(136,200))
            #img = img[40:142, 16:120]
            #img = cv2.imwrite("./raw/"+str(count)+".jpg",img)            
            count += 1
        else:
            pass
#             f=open("label.txt","a")
#             f.write(item[1:]+"\n")
#             f.close()
    print('Finish')
```
